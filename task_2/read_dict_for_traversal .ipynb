```python
import json
```

```python
def readFile(filename):
    dict_list=[]
    with open(filename,'r',encoding='utf8') as f:
        for line in f.readlines():
            dict_list.append(line.strip())
    return dict_list
```

```python
def saveDict(dic,filename):
    with open(filename,'w',encoding='utf8') as f:
        f.write(str(dic))
```

```python
def traversal(dict_):
    '''遍历一个字典'''
    global count
    for k,v in dict_.items():
        if isinstance(v,dict):#如果值是字典，递归遍历
            traversal(v)
        else:#否则访问
            print(k,':',v)
            count+=1
```

```python
def traversal_1(dict_,list_re):
    '''遍历一个字典，并重新赋值'''
    global count#使用全局变量，因为涉及到递归
    for k,v in dict_.items():
        if isinstance(v,dict):#如果值是字典，递归遍历
            traversal_1(v,list_re)
        else:#否则访问赋值
            dict_[k]=list_re[count]
            count+=1
    return dict_
```

```python
def main():
    filename1='demo1.txt'#regex
    filename2='demo2.txt'#dcit
    filename3='demo3.txt'#dict
    re_list=readFile(filename1)#读取正则文件到列表
    dict_list=readFile(filename2)#读取字典到列表
    for dic in dict_list:
        dic=eval(dic)#解析字典
        dic=traversal_1(dic,re_list)#遍历赋值
        with open('demo4.txt','w',encoding='utf8') as f:
            json.dump(dic,f,ensure_ascii=False)#ensure_ascii=False确保输出的中文可以正常显示
        saveDict(dic,filename3)#保存字典
```

```python
global count#声明全局变量
count=0
main()
```
